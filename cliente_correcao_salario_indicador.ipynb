```python
df_ipca = pd.read_csv('../../git/ibge-indicadores-censitarios/dados/indices_de_correcao/IPCA.csv', sep=';', decimal=',')
```

```python
df_ipca
```

```python
from multiprocessing import Pool
import numpy as np

def compara(df):
    df_vazio=pd.DataFrame(columns=["taxa_acum"])
    #itera df_sicli
    for i in df.values:
        array = 1
        #itera salarios_minimos
        for j in df_ipca.values:
            if int(j[0]) >= int(i[10]):
                if j[1] != 0.0:
                    array = array*(1+j[1]/100)
        df_vazio = df_vazio.append({'taxa_acum':array},ignore_index=True)
    return df_vazio

def parallelize_compare(df, func, n_cores=12):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split)).reset_index(drop=True)
    pool.close()
    pool.terminate()
    pool.join()
    return df


start = time.time() # inicia o tempo de execução

df_full = pd.concat([df_ren, parallelize_compare(df_ren,compara,n_cores=4)], axis=1)

elapsed = time.time() - start # informa o tempo de execução
elapsed / 60
```
