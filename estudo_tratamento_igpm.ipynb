```python
df_igpm = pd.read_csv(arquivo, sep=sep, decimal=',')
```

```python
df_sicli = pd.read_csv("../dados/notebook/cliente_area_de_ponderacao/cliente_area_de_ponderacao.csv", delimiter=';')
```

```python
df_sicli
```

```python
df_sicli = df_sicli[(df_sicli['vr_renda_bruta'] != 0) & (df_sicli['dt_apuracao_renda'] != "?")]
```

```python
df_sicli = df_sicli[['nu_pessoa_p17', 'vr_renda_bruta','dt_apuracao_renda']]
```

```python
df_sicli['dt_apuracao_renda'] = df_sicli['dt_apuracao_renda'].replace(regex=r'\-', value='')
df_sicli['dt_apuracao_renda'] = (df_sicli['dt_apuracao_renda'].astype(int) / 100).astype(int)
```

```python
df_sicli
```

```python
df_igpm.info()
```

```python
array_full = []
for i in df_sicli.values:
    array = 1
    for j in df_igpm.values:
        if int(j[0]) >= int(i[2]):
            if j[1] != 0.0:
                array = array*(1+j[1]/100)
    array_full.append(array)
```

```python
df_sicli['pc_correcao_salario_igpm'] = array_full
```

```python
df_sicli['vr_salario_corrigido_igpm'] = round(df_sicli['vr_renda_bruta'] * df_sicli['pc_correcao_salario_igpm'] , 2)
```
